# PowerCo Customer Churn - Exploratory Data Analysis

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
pd.set_option('display.max_columns', None)
```

```
client_df = pd.read_csv('client_data.csv')
price_df = pd.read_csv('price_data.csv')

print('Client Data Shape:', client_df.shape)
print('Price Data Shape:', price_df.shape)
```

```
for col in client_df.columns:
    print(col, client_df[col].nunique())
```

```
date_cols = ['date_activ','date_end','date_modif_prod','date_renewal']

for col in date_cols:
    if col in client_df.columns:
        client_df[col] = pd.to_datetime(client_df[col])
```

```
sns.countplot(data=client_df, x='churn')
plt.title('Churn Distribution')
plt.show()
```

```
sns.histplot(client_df['cons_12m'], bins=50, kde=True)
plt.title('Annual Consumption Distribution')
plt.show()
```

```
sns.histplot(client_df['net_margin'], bins=50, kde=True)
plt.title('Net Margin Distribution')
plt.show()
```

```
sns.histplot(client_df['num_years_antig'], bins=20, kde=True)
plt.title('Customer Tenure')
plt.show()
```

```
client_df['channel_sales'].value_counts().head(10).plot(kind='bar')
plt.title('Top Sales Channels')
plt.show()
```

```
sns.countplot(data=client_df, x='has_gas')
plt.title('Gas Subscription')
plt.show()
```

```
pd.crosstab(client_df['has_gas'], client_df['churn'])
```

```
price_cols = ['price_off_peak_var','price_peak_var','price_mid_peak_var']

for col in price_cols:
    if col in price_df.columns:
        sns.histplot(price_df[col], bins=50, kde=True)
        plt.title(col)
        plt.show()
```

```
plt.figure(figsize=(14,10))
sns.heatmap(client_df.select_dtypes(include=np.number).corr(), cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

## Key Findings

1. Churned customers represent a minority of the dataset.
2. Energy consumption variables are heavily right-skewed.
3. Several variables contain significant outliers.
4. Customer tenure varies substantially across customers.
5. Pricing variables show variation that may influence churn behaviour.
6. Further feature engineering is required to quantify price sensitivity.
7. Additional investigation will compare pricing differences between churned and retained customers.
